```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('Day_28_Clustering_Dataset.csv')

# Display the first five rows
print(data.head())

# Scatter plot of Feature_1 vs Feature_2
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Feature_1', y='Feature_2')
plt.title('Feature_1 vs Feature_2')
plt.xlabel('Feature_1')
plt.ylabel('Feature_2')
plt.show()

# Compute basic statistical measures
print(data.describe())
```

```python
from sklearn.cluster import KMeans

# Applying K-Means clustering with K=3
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(data[['Feature_1', 'Feature_2']])

# Plot the clustered data
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='Feature_1', y='Feature_2', hue='Cluster', palette='Set1')
plt.title('K-Means Clustering (K=3)')
plt.xlabel('Feature_1')
plt.ylabel('Feature_2')
plt.legend()
plt.show()
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data[['Feature_1', 'Feature_2']])
    wcss.append(kmeans.inertia_)

# Plotting the Elbow Curve
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('WCSS')
plt.show()
```

```python
for k in [2, 4, 5]:
    kmeans = KMeans(n_clusters=k, random_state=42)
    data['Cluster'] = kmeans.fit_predict(data[['Feature_1', 'Feature_2']])

    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=data, x='Feature_1', y='Feature_2', hue='Cluster', palette='Set1')
    plt.title(f'K-Means Clustering (K={k})')
    plt.xlabel('Feature_1')
    plt.ylabel('Feature_2')
    plt.legend()
    plt.show()
```
